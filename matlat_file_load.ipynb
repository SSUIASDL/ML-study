```python
from scipy import io

mat_file=io.loadmat('35x35_50000Hz_Nt1001011_CFRP_2mm.mat')
mat_file.keys() # 딕셔너리 데이터를 이야기... mat file에서 ... 해당 내용들을 보고 싶을때 사용
mat_file["Nx"]
print(mat_file["Nx"])
stc=mat_file["stc"]
```

```python
stc.shape
```

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.imshow(stc.real)
plt.xlabel("x"), plt.ylabel("y")
plt.title("Steady state response (real)")

plt.subplot(1,2,2)
plt.imshow(stc.imag), plt.xlabel("x")
plt.title("Steady state response (imag)")
```
